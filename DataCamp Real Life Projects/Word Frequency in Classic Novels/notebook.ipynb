```python
# Use this cell to begin your analysis, and add as many as you would like!
from bs4 import BeautifulSoup as bs
import requests
from nltk.tokenize import regexp_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import collections
import pandas as pd
```

```python
r = requests.get('https://www.gutenberg.org/files/16/16-h/16-h.htm')
soup = bs(r.text)
text = soup.get_text()
```

```python
pattern = '\w+'
tokens = regexp_tokenize(text, pattern=pattern)
len(tokens)
```

```python
stop_words = set(stopwords.words('english'))
tokens_cleaned = [x.lower() for x in tokens]
tokens_cleaned = [x for x in tokens if x not in stop_words]
len(tokens_cleaned)
```

```python
counted = collections.Counter(tokens_cleaned)
word_freq = pd.DataFrame(counted.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
```
